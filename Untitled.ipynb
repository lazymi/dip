```python
from google.colab.patches import cv2_imshow
import cv2
img=cv2.imread("/content/f1.jpg")
cv2_imshow(img)
```

```python
g=cv2.GaussianBlur(img,(5,5),0)
cv2_imshow(g)
```

```python
edge=cv2.Canny(img,100,200)
cv2_imshow(edge)
```

```python
import numpy as np
lines = cv2.HoughLinesP(edge, 1, np.pi/180, 60, np.array([]), 50, 5)
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(img, (x1, y1), (x2, y2), (20, 220, 20), 3)
cv2_imshow(img)
```
